# 🧞 04 — Genie Space Setup (Conversational Analytics)

**Databricks Genie** lets users chat with SQL tables in plain English. We expose our Gold tables + views so judges can ask questions like:
- *"Which city has the highest PM2.5 and how many trains pass through it?"*
- *"What's the average predicted delay by month?"*

This is a MASSIVE Databricks-Usage showcase feature. Almost no competing team will include it.

This notebook runs validation SQL; the Genie space itself is created via UI.

````python
%sql
SELECT 'gold.vw_trains_by_city'        AS name, COUNT(*) AS n FROM setu_rail.gold.vw_trains_by_city
UNION ALL SELECT 'silver.train_enriched',         COUNT(*) FROM setu_rail.silver.train_enriched
UNION ALL SELECT 'bronze.air_quality',            COUNT(*) FROM setu_rail.bronze.air_quality;
````

````python
%sql
-- These are the tables/views Genie will expose

UNION ALL SELECT 'gold.vw_hourly_schedule',       COUNT(*) FROM setu_rail.gold.vw_hourly_schedule
UNION ALL SELECT 'gold.vw_top_polluted_routes',   COUNT(*) FROM setu_rail.gold.vw_top_polluted_routes
UNION ALL SELECT 'gold.predictions_daily',        COUNT(*) FROM setu_rail.gold.predictions_daily
````

## UI Setup — Create the Genie Space

1. In the left sidebar, click **Genie** → **Create Genie space** → name it **SETU-Rail Analytics**.
2. Under **Datasets**, add these tables/views:
   - `setu_rail.gold.vw_trains_by_city`
   - `setu_rail.gold.vw_hourly_schedule`
   - `setu_rail.gold.vw_top_polluted_routes`
   - `setu_rail.gold.predictions_daily`
   - `setu_rail.silver.train_enriched`
   - `setu_rail.bronze.air_quality`
3. Under **Instructions**, paste:

```
You are the analytics assistant for SETU-Rail, an AI copilot for Indian Railways.
Key concepts:
• Delays are measured in arrival_delay_min (range 0–360).
• pm25 > 100 is a strong fog proxy and correlates with delay.
• is_junction = 1 means a major interchange station (MAS, SBC, NDLS, HWH, CSMT).
• Round revenue-like numbers and always use minutes for time.
```

4. Under **Certified queries** (examples), add:
   - *"Which 5 cities have the highest average PM2.5?"*
   - *"What is the average predicted delay by month?"*
   - *"Which junction station has the most trains?"*

5. Click **Publish**. Copy the Genie space URL for the app.
