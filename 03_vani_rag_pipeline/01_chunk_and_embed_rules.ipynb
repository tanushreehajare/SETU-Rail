# 💬 01 — Vani: Validate Rulebook Chunks for Vector Search

The chunks already exist in `setu_rail.silver.rules_chunks` (built in `01_dhara/03_ingest_rulebook_pdfs`). This notebook validates them and prepares for **Databricks Vector Search** ingestion.

Vector Search auto-generates embeddings using the `databricks-bge-large-en` endpoint — no manual embedding work needed. This is a HUGE Databricks-depth win over manual FAISS.

```python
%sql
-- Validate chunks table & enable Change Data Feed (required by Vector Search sync)
SELECT COUNT(*) AS total_chunks FROM setu_rail.silver.rules_chunks;

-- Confirm CDF is on
SHOW TBLPROPERTIES setu_rail.silver.rules_chunks;
```

```python
%sql
-- If for some reason CDF is off, turn it on:
ALTER TABLE setu_rail.silver.rules_chunks
SET TBLPROPERTIES (delta.enableChangeDataFeed = true);
```

```python
%sql
-- Peek at interesting sections
SELECT id, source, page, section, SUBSTRING(text, 1, 180) AS preview
FROM   setu_rail.silver.rules_chunks
WHERE  text RLIKE '(?i)refund|compensation|untoward|accident|passenger|ticket|delay'
LIMIT 10;
```
